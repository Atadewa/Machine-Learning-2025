### Kegunaan Library

1. PyPREP
    - Digunakan untuk preprocessing sinyal EEG (Electroencephalogram).
    - Fungsinya membersihkan noise/artefak sehingga sinyal otak bisa lebih akurat untuk analisis.

2. Scipy
    - Library dasar untuk komputasi ilmiah.
    - Dipakai untuk matematika, statistik, optimasi, dan signal processing (misalnya filtering, FFT).

3. wandb (Weights & Biases)
    - Alat untuk experiment tracking pada Machine Learning/Deep Learning.
    - Membantu mencatat hasil training, metrik (akurasi, loss), dan membandingkan eksperimen.

4. pyECG
    - Digunakan untuk analisis sinyal ECG (Electrocardiogram).
    - Fungsinya mendeteksi detak jantung (R-peak), menghitung HRV, dan pemrosesan sinyal jantung.
